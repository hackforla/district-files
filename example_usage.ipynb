# Example: Working with LA District Boundaries

This notebook shows how to open and use the data saved in two formats:
- Raw ArcGIS JSON
- Standard GeoJSON


```python
import json
import geopandas as gpd
import matplotlib.pyplot as plt
```

## 1. Load the raw ArcGIS JSON

```python
with open("shapefiles_output/assembly.json", "r", encoding="utf-8") as f:
    raw_json = json.load(f)

print("Top-level keys in ArcGIS JSON:", raw_json.keys())
print("Number of features:", len(raw_json.get("features", [])))

# Peek at one feature's attributes
if raw_json.get("features"):
    print("Example attributes:\n", raw_json["features"][0]["attributes"])
```

## 2. Load the GeoJSON with GeoPandas

```python
gdf = gpd.read_file("shapefiles_output/assembly.geojson")

print("GeoDataFrame info:")
print(gdf.info())
```

## 3. Quick Plot

```python
gdf.plot(figsize=(8, 6), edgecolor="black", facecolor="lightblue")
plt.title("Assembly Districts (from GeoJSON)")
plt.show()
```

## 4. Access attribute data

```python
print("First 5 rows of attribute data:")
print(gdf.head())
```
